
# Data Preparation

As mentioned in section 2.4 of the Iteration 4 document, there is some pre-processing of the original data files (TABLECODE7361_ALL_Data_Children and young people charged in court - most serious offence calendar year.csv; TABLECODE7362_ALL_Data_Children and young people given an order in court - most serious offence calendar year.csv).

The tasks below were processed using R. This was installed on the EC2 server from Ubuntu using the following command: sudo apt install r-base-core 

The pre-processing tasks that I employed R to perform on EC2 are as follows: 
(1) All instances of the Value field that are set to 3 are to be set to 1 to counteract SNZ’s rounding policy. 
(2) Each line will be duplicated by the number of times indicated in its Value field to output one row in the data for every actual case that matches that specification. 
(3) Part of the data in table 7361 – Children and Youths charged needs to be merged with table 7362 – Child and youth sentences. A select on table 7361 is used to choose only those observations that do not pertain to children and youth who were sentenced in court (which is in dataset 7362). The desired items from 7361 satisfy the following select statement:
(Outcome = "Youth Court proved (absolute discharge under s282)" or Outcome = "Dismissed" or Outcome = "Withdrawn" or Outcome = "Other outcome")
(4) Commas in the “Outcome” (and/or “Order”) columns should not have commas. 

The R code that was used to perform these tasks in R on EC2 is available from the GitHub in the file: “Pre-processing_SNZ_7361_7362.R”.

```python
# The merged data file generated on EC2 using R that is described above
# will now be read. The Flags column will be removed.
pandas_merged_df = pd.read_csv('Datasets/child_youth_stats_repeated.csv', index_col=0)
pandas_merged_df = pandas_merged_df.drop(['Flags'], axis=1)
pandas_merged_df
```


# Data Transformation

There are four data scenarios requiring the transformation of two data frames:
scenario (0) court order data for those charges proved (table 7362) with outcomes ungrouped
scenario (1) merged data (court order and all charges) with outcomes grouped
scenario (2) merged data (court order and all charges) with both outcomes and years grouped
scenario (3) court order data for those charges proved (table 7362) with outcomes grouped

Here we add two extra columns: YearGroup and OutcomeSeverity

Year Group 1: 1992, 1993, 1994, 1995, 1996, 1997

Year Group 2: 1998, 1999, 2000, 2001, 2002

Year Group 3: 2003, 2004, 2005, 2006, 2007

Year Group 4: 2008, 2009, 2010, 2011, 2012

Year Group 5: 2013, 2014, 2015, 2016, 2017


```python
# Create YearGroup column for merged data
bins = [1, 1997, 2002, 2007, 2012, 2017]
names = [1, 2, 3, 4, 5]

pandas_merged_df['YearGroup'] = pd.cut(pandas_merged_df.Year, bins, labels=names)
pandas_merged_df.sample(20)
```
